```python
s_cylinder = occiput.DataSources.Synthetic.Shapes.uniform_cylinder(voxels=pet.activity_shape, axis=2, radius=0.12, inner_value=2.5, length=100, center=[0.498,0.498,0.5])
plot(ars[:,172,64]), axis([0,344,0,20])
hold(True)
plot(s_cylinder.data[:,172,64],'r'), axis([0,344,0,3])
```

```python
j = 0
nn = 126
p_sum = zeros([344,344])
p_a = zeros([344,344])
ss_cylinder = zeros([344,344])
aaa = 0

for i in range(nn): 
    j = j+1
    ss_cylinder = s_cylinder.data[:,:,j]
    aa = numpy.count_nonzero(ss_cylinder)
    aaa = aaa + aa
    p_a = ars[:,:,j]
    p_sum[ss_cylinder > 0] = p_sum[ss_cylinder > 0] + p_a[ss_cylinder > 0]
    
v_sum = p_sum.sum() / aaa
```

```python
figure(figsize=[20,9])
imshow(p_sum[110:232,110:232], vmax = 250)
```

```python
j = 10
nn = 100
p_a = zeros([344,344])
p_b = zeros([344,344])
ss_cylinder = zeros([344,344])
p_values = zeros([nn,])
aa = 0

for i in range(nn): 
    j = j+1
    ss_cylinder = s_cylinder.data[:,:,j]
    aa = numpy.count_nonzero(ss_cylinder)
    p_a = ars[:,:,j]
    p_b[ss_cylinder > 0] = p_a[ss_cylinder > 0]
    p_values[i] = p_b[:,:].sum() / aa
```

```python
p_values_norm = ((p_values - v_sum) / v_sum) * 100
plot(p_values_norm)
```
