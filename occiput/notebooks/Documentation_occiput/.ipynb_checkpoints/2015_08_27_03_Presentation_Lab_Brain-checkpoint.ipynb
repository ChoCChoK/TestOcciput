ITERATIVE RECONSTRUCTION 

```python
def estimate_activity_rs(self, attenuation, iterations=10, epsilon=1e-8, subsets_matrix=None): 
        import time
        
        pix_x = 2.08626
        pix_y = 2.08626
        pix_z = 2.03125

        from occiput.Reconstruction.PET import ROI
        roi = ROI((0.5*pet.activity_size[0]-pix_x/2,0.5*pet.activity_size[1]-pix_y/2,0.5*pet.activity_size[2]-0.5*pix_z,0,0,0))        
#       MIND THAT THE SIGN OF -0.5*pix_z IS PROBABLY CORRECT ???
        if subsets_matrix is None: 
            subsets_matrix = ones([11,252],dtype=float32,order="C")
            
        activity   = ones(self.activity_shape,dtype=float32,order="F")
        
        sensitivity = self.sensitivity.data
        attenuation = attenuation.to_nd_array()
        
#        attenuation = ones(attenuation.shape,dtype=float32,order="C") ###
        
        sens_total  = ones(self.prompts.to_nd_array().shape,dtype=float32,order="C")
        measurement = self.prompts.to_nd_array() 
        att_sens = attenuation * sensitivity
        randoms     = self.randoms.data / (att_sens + epsilon) 
        scatter     = (self.scatter.data + epsilon) / (attenuation + epsilon)
        
#        randoms = 0*randoms ###
#        scatter = 0*scatter ###
        
        norm = self.backproject(att_sens, subsets_matrix=subsets_matrix, roi_activity=roi) 

        for i in range(iterations): 
            proj = self.project(activity, subsets_matrix=subsets_matrix, roi_activity=roi)
            update = (self.backproject( measurement/(proj.to_nd_array() + randoms + scatter + epsilon), subsets_matrix=subsets_matrix, roi_activity=roi ).data+epsilon) / (norm.data + epsilon) 

            mask = self.get_activity_mask()
            if mask is not None: 
                activity = activity * update * mask.data
            else: 
                activity = activity * update 

            print "Iteration: %d / %d"%(i+1,iterations), time.ctime(); time.sleep(0.2)
        return activity
```

```python
T_sec = (pet.prompts.get_time_end()-pet.prompts.get_time_start())/1000.0
pet.set_projection_scale(8.58341777027e-05 * T_sec * 0.859541773948) 

pet.binning.angles_axial = float32( linspace(0,pi-pi/252.0,252) )
```

```python
subsets_matrix = zeros([11,252],dtype=float32,order="C")
#subsets_matrix[range(11),:]=1
subsets_matrix[:,:]=1

iterations = 40
ars = estimate_activity_rs(pet, attenuation=attenuation, iterations=iterations, subsets_matrix=subsets_matrix)
```
